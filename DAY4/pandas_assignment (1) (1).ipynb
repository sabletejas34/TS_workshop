```python
import pandas as pd
import numpy as np
df = pd.read_csv("Titanic-Dataset.csv")
df.head()

```

```python
df.index
```

```python
#1. Find total number of survived(0) and non-survived(1)

df['Survived'].value_counts()
```

```python
#2. Find max and min fairprice of each Pclass (1 and 2 and 3)
d=df[['Pclass','Fare']]
print(d)


print('min fairprice of each Pclass is')
min_val= df.groupby("Pclass")['Fare'].min()
print(min_val)
print('*****************')
print('\n max fairprice of each Pclass is')
max_val= df.groupby("Pclass")['Fare'].max()
print(max_val)
```

```python
#3. Handle missing values in age and cabin columns
a=(df[['Age','Cabin']])
print(a)
a.isna().sum()
```

```python
a.fillna(value='FILL VALUE')
```

```python
#4. Find average fair price of lower class people (Pclass = 3) who survived
e=df[(df['Survived'] ==1) & (df['Pclass'] ==3)]
f=e['Fare'].mean()
print(f)


```

```python
#5. Learn dna implement how to find duplicate columns and rows and how to remove them
f=df['Fare'].nunique()
print(f)
T = df.drop_duplicates(subset=['Fare'], keep='first')
T['Survived'].value_counts()
```
