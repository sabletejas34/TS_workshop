```python
#find common elemnts from list l1 and l2
l1 = [1,2,3,4]
l2 = [3,4,5,6]
import numpy as np
l3=np.intersect1d(l1,l2)
print(l3)
```

```python
d1 = {'a': 1, 'b': 2, 'c': 3, 'g': 4}
d2 = {'a': 7, 'f': 2, 'g': 3, 'h': "101"}

#find common key from d1 and d2
print("common keys are:" )
for key in d1.keys():
    if key in d2.keys():
        print(key,end=" ")
print()
#find common value from d1 and d2
print("common values are:" )
for value in d1.values():
  if value in d2.values():
    print(value,end=" ")
```

```python
#replace the values of d1 to their negatie value
for key, value in d1.items():
   d1[key] =-abs(value)
print(d1)
```

```python
#replace key to value and value to key in d2
print(d2)
d3 = {}
for key, value in d2.items():
    d3[value] =key
print(d3)
```
